```python
import os
import re
from pprint import pprint
from langchain.document_loaders import PyPDFLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.vectorstores import faiss
from langchain.embeddings import HuggingFaceEmbeddings
```

```python
# Загружаем книги в loader
```

```python
dir = '/content/sample_data'
documents = []
for root, dirs, files in os.walk(dir):
  for file in files:
    if file.endswith(".pdf"):
      print(f'root={root} file={file}')
      loader = PyPDFLoader(os.path.join(root, file))
      documents.extend(loader.load())
```

```python
print(type(documents))
print(len(documents))
print(documents[12].metadata)
print(documents[12].page_content)
```

Разделение на chunks

```python
text_splitter = RecursiveCharacterTextSplitter(chunk_size = 1024, chunk_overlap = 0)
source_chunks = text_splitter.split_documents(documents)
print(type(source_chunks))
print(len(source_chunks))
print(source_chunks[1000].metadata)
print(source_chunks[1000].page_content)
```

Embeddings

```python
model_id = 'intfloat/multilingual-e5-large'
model_kwargs = {'device': 'cuda'}
embeddings = HuggingFaceEmbeddings(
    model_name = model_id,
    model_kwargs = model_kwargs
)
```

Векторная база знаний

```python
db = faiss.FAISS.from_documents(source_chunks, embeddings)
```

Similarity search

```python
topic =  'Москва'
docs = db.similarity_search(topic, k = 3)
# message_content = re.sub(r'\n{2}', ' ', '\n'.join([f'\n#### {i+1} Relevant chunk ####\n' + str(doc.metadata) + '\n']))
print(docs[0])
```
